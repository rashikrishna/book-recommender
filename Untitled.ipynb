# Building 'User-User Collaborative Filtering' Based Recommender

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#print(cosine_similarity(y,y))
```

```python
# to downsize data : only including users who have rated more than 10 books]
pd.DataFrame(x['userid'].unique()).count()
```

```python
users10 = x.groupby('userid').count()
users10 = users10.loc[users10.isbn>=100]
users10=users10.index
```

```python
x=x.loc[x.userid.isin(users10)]
x
```

```python
x.groupby('userid').count()
```

```python
#including books with more than 10 ratings
books10=x.groupby('isbn').count()
books10=books10.loc[books10.bookrating>=10]
books10=books10.index
books10
```

```python
x = x.loc[x.isbn.isin(books10)]
x
#x contains only users which have rated more than 10 books and books which have more than 10 ratings
```

```python
y=x.pivot_table(columns='isbn',values='bookrating',index='userid')
y
#y is the explicit_matrix
```

```python
mean_rating = pd.Series(y.mean(axis=1))
mean_rating
#mean rating of each user
```

```python
#y = y.fillna(0)
y=y.sub(pd.Series(y.mean(axis=1)),axis='index')
y= y.fillna(0)
```

```python
y
```

```python
similarity_matrix=pd.DataFrame(cosine_similarity(y,y),columns=y.index,index=y.index)
similarity_matrix
```

```python
def predictrating(userid,bookid,similarity_mat,explicit_rating):
    similarity = similarity_mat.loc[userid,:]
    rating = explicit_rating[bookid]
    rating = rating.mul(similarity,axis="index")
    rating = rating.sum()/(similarity.sum()-1)
    rating = rating+mean_rating.loc[userid]
    if rating>10:
            rating=10
    elif rating<0:
            rating=0
    return rating
```

```python
import operator
def recommendbook(userid,similarity_mat,explicit_rating,books_table=books):
    users_matrix = explicit_rating.index
    if userid not in users_matrix:
        print("No User found")
        return
    predicted_books = []
    for i in range(explicit_rating.shape[1]):
        rat = predictrating(userid,explicit_rating.columns[i],similarity_mat,explicit_rating)
        if rat>5 and explicit_rating.loc[userid,explicit_rating.columns[i]]==0:
            #print(rat)
            predicted_books.append([explicit_rating.columns[i],rat])
    predicted_books.sort(key=operator.itemgetter(1),reverse=True)
    top_five = []
    for i in range(10):
        top_five.append(predicted_books[i])
    
    top_five =  np.array(top_five)
    return list(books_table.loc[books_table.isbn.isin(top_five[:,0])]['booktitle'])
    #return top_five
```

```python
recommendbook(2033,similarity_matrix,y,books)
```
